```python
df.columns
df.drop(['prices.amountMax', 'prices.amountMin'], axis=1, inplace=True)
```

```python
df.head()
```

```python
   
# X=df.drop(['labels'], axis=1)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

brand = le.fit_transform(df['brand'])
name = le.fit_transform(df['name'])
color = le.fit_transform(df['prices.color'])
isSale= le.fit_transform(df['prices.isSale'])
size= le.fit_transform(df['prices.size'])

df['brand']=brand
df['name']=name
df['prices.color']=color
df['prices.isSale']=isSale
df['prices.size']=size  	

df.info()

```

# Train-test splitting

```python
from sklearn.model_selection import train_test_split 
X=df.drop('labels', axis=1)
y=df['labels'].copy()
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

# Implementing a model

```python
from sklearn.linear_model import LinearRegression
# from sklearn.tree import DecisionTreeRegressor
# from sklearn.ensemble import RandomForestRegressor
model=LinearRegression()
model.fit(X_train, y_train)
predictions=model.predict(X_test)

d=pd.DataFrame(predictions)
d
```

```python

from sklearn.metrics import mean_squared_error
final_mse=mean_squared_error(y_test,predictions)
# final_rmse=np.sqrt(final_mse)
# final_rmse
final_mse
```
